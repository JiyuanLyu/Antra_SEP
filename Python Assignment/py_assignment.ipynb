```python
# Problem 1
import pandas as pd

# open the file people_1
people_1 = pd.read_csv('people_1.txt')
people_1[['FirstName', 'LastName', 'Email', 'Phone', 'Address']] = people_1['FirstName\tLastName\tEmail\tPhone\tAddress'].str.split('\t', expand=True)
people_1.drop('FirstName\tLastName\tEmail\tPhone\tAddress', inplace=True, axis=1)

people_1.head()
```

```python
# open the file people_2
people_2 = pd.read_csv('people_2.txt')
people_2[['FirstName', 'LastName', 'Email', 'Phone', 'Address']] = people_2['FirstName\tLastName\tEmail\tPhone\tAddress'].str.split('\t', expand=True)
people_2.drop('FirstName\tLastName\tEmail\tPhone\tAddress', inplace=True, axis=1)

people_2.head()
```

```python
# Combine two files into one
people = people_1.append(people_2, ignore_index=True)
people
```

```python
# Convert the first name and last name into same format
from ast import Lambda

people["FirstName"] = people["FirstName"].apply(lambda x: x.capitalize().replace(" ", ""))
people["LastName"] = people["LastName"].apply(lambda x: x.capitalize().replace(" ", ""))
people
```

```python
# Formatting the email addresss
people["Email"] = people["Email"].apply(lambda x: x.replace(" ", ""))
people
```

```python
# Formatting the phone number
# Write a function for formatting
def format_phone(num):
    if len(num) == 10:
        sign = num.find('-')
        num = num[0:3] + '-' + num[3:6] + '-' + num[6:10]
    return(num)

people["Phone"] = people["Phone"].apply(lambda x: format_phone(x))
people
```

```python
# Formatting the address
people["Address"] = people["Address"].apply(lambda x: x.translate({ord(c): None for c in "No.#"}))
people
```

```python
# drop duplicate rows and reindex
import numpy as np

df = people.drop_duplicates(inplace=False)
df.index = np.arange(0, len(df))
df
```

```python
# Export a CSV file
df.to_csv('people.csv', index=False)
```

```python
# Problem 2
# read json file
import json
with open("movie.json", 'r') as j:
    movies = json.load(j)
    # Since the length is 9995, 
    # split it into 8 file
    # each file have 1250 object

    for i in range(0, 8):
        filename = 'movie_'+str(i)+'.json'

        with open(filename, 'w') as out_json_file:
            # Save each obj to their respective filepath
            # with pretty formatting thanks to `indent=4`
            new = {"movie": movies["movie"][(0+i*1250):(1250+i*1250)]}
            json.dump(new, out_json_file, indent = 4)
```
