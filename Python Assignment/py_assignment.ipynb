```python
# Question 1
import pandas as pd

# open the file people_1
people_1 = pd.read_csv('people_1.txt')
people_1[['FirstName', 'LastName', 'Email', 'Phone', 'Address']] = people_1['FirstName\tLastName\tEmail\tPhone\tAddress'].str.split('\t', expand=True)
people_1.drop('FirstName\tLastName\tEmail\tPhone\tAddress', inplace=True, axis=1)

people_1.head()
```

```python
# open the file people_2
people_2 = pd.read_csv('people_2.txt')
people_2[['FirstName', 'LastName', 'Email', 'Phone', 'Address']] = people_2['FirstName\tLastName\tEmail\tPhone\tAddress'].str.split('\t', expand=True)
people_2.drop('FirstName\tLastName\tEmail\tPhone\tAddress', inplace=True, axis=1)

people_2.head()
```

```python
# Combine two files into one
people = people_1.append(people_2, ignore_index=True)
people
```

```python
# Convert the first name and last name into same format
from ast import Lambda

people["FirstName"] = people["FirstName"].apply(lambda x: x.capitalize().replace(" ", ""))
people["LastName"] = people["LastName"].apply(lambda x: x.capitalize().replace(" ", ""))
people
```

```python
# Formatting the email addresss
people["Email"] = people["Email"].apply(lambda x: x.replace(" ", ""))
people
```

```python
# Formatting the phone number
# Write a function for formatting
def format_phone(num):
    if len(num) == 10:
        sign = num.find('-')
        num = num[0:3] + '-' + num[3:6] + '-' + num[6:10]
    return(num)

people["Phone"] = people["Phone"].apply(lambda x: format_phone(x))
people
```

```python
# Formatting the address
people["Address"] = people["Address"].apply(lambda x: x.translate({ord(c): None for c in "No.#"}))
people
```

```python
# drop duplicate rows and reindex
import numpy as np

df = people.drop_duplicates(inplace=False)
df.index = np.arange(0, len(df))
df
```

```python
# Export a CSV file
df.to_csv('people.csv', index=False)
```

```python
# Question 2
# read json file
import json
with open("movie.json", 'r') as j:
    movies = json.load(j)
    # Since the length is 9995, 
    # split it into 8 file
    # each file have 1250 object

    for i in range(0, 8):
        filename = 'movie_'+str(i)+'.json'

        with open(filename, 'w') as out_json_file:
            # Save each obj to their respective filepath
            # with pretty formatting thanks to `indent=4`
            new = {"movie": movies["movie"][(0+i*1250):(1250+i*1250)]}
            json.dump(new, out_json_file, indent = 4)
```

# Question 3

**SaaS**, Software as a Service, uses the web to deliver applications managed by a third-party vendor, and it can be accessed on the clients' side. The users can benefit from the functions without downloading or installing the application. 

**PaaS**, Platform as a Service, is primarily used as hardware and software tools on the internet. It's usually used for application development and other development, while providing cloud components to software. Users can run their own copies of applications with the cloud provider's infrastructure, without worrying about the storage spaces and maintaining servers or hard disks. 

**IaaS**, Infrastructure as a Service, provides service for accessing, monitoring, and managing remote data center infrastructures. The clients do not have to purchase the hardware outright and use IaaS for their needs. IaaS works primarily with cloud-based and provide services such as storage, networking, and virtualization.

The difference between these three services is that PaaS, is a variant of SaaS while the clients can run their own copies of the application, using the cloud provider’s infrastructure. And compared to SaaS and PaaS, IaaS users are responsible for managing applications, data, runtime, middleware, and OSes. However, Providers still manage virtualization, servers, hard drives, storage, and networking. 


# Question 4

ETL means the Extract, Transform, and Load process for data. ELT means Extract, Load, and Transform process for data. ETL transforms data on a separate processing server and does not transfer raw data into the data warehouse, while ELT transforms data within the data warehouse and sends raw data directly to the data warehouse. In the transform processing, ELT moves data from the data source to staging into the data warehouse and cleans sensitive and secure data before loading it into the data warehouse while ELT does not. ELT can perform complex data transformation and be more cost-effective than ELT. 

|| ETL | ELT |
|-----|-----|-----|
|Transformation Speed|ETL is time insensitive. The data must be transformed before loading.|Faster than ETL, can load and transform at the same time|
|Code-Based Transformation|Best for compute-intensive transformation & pre-cleaning.|Speed & efficient.|
|Privacy|Pre-load transformation can eliminate PII (helps for HIPPA).|Need more privacy safeguard.|
|Data Lake|Do not have data lake compatibility.|Have Data Lake compatibility.|
|Data Structure|Structured|It could be structured, semi-structured, unstructured.|
|Cost|Separate servers can create cost issues.|Cost Less.|
|Data Volume|For small datasets with complicated transformations.|For large datasets that require speed and efficiency.|
